#NESTED DICTIONARIES

```python
student = {
    "name":"Ramcharan",
    "age":26,
    "marks":{
        'maths':98,
        'science':89,
        'social':100,
        'hindi':99,
        'biology':{
            "sub1":"Natural science",
            'physics':'Atomic physics'
        }
    }
}
print(student)
len(student)
```

```python
print(student['marks']['biology']['physics'])
```

```python
students = {
    "student1": {
        "name": "Rahul",
        "age": 21,
        "marks": {
            "math": 80,
            "science": 75
        }
    },
    "student2": {
        "name": "Anita",
        "age": 22,
        "marks": {
            "math": 90,
            "science": 85
        }
    }
}
print(students)
```

```python
for stud1 , sub1  in students['student1'].items():
  print(stud1,":", sub1)
```

```python
#Important pattern for AIML and it is industrial standards

for student_id , details in students.items():
  print("------------------------")
  print("Name" ,":", details["name"])

  for subject , marks in details['marks'].items():
    print(subject,":", marks)
```

```python
for student_names , name in students.items():
  print("Names",":", name['name'])
  print("Math", "=", name['marks']['math'])
```
